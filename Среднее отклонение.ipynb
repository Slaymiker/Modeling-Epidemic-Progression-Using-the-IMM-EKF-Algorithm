```python
import os
import pandas as pd
from glob import glob

# Списки окон обучения и прогноза
TRAIN_WINDOW_lst = [3, 5, 7, 10, 15, 20, 25, 30]
PREDICT_WINDOW_lst = [3, 5, 7, 10, 15, 20, 25, 30]

# Имена столбцов, по которым считаем средние
col_names = ["% откл S", "% откл I", "% откл R"]

# Рабочая папка (где выполняется код)
work_dir = os.getcwd()

for train_window in TRAIN_WINDOW_lst:
    for predict_window in PREDICT_WINDOW_lst:
        # формируем строку окна, например "5_5"
        wind = f"{train_window}_{predict_window}"

        print("\n" + "=" * 80)
        print(f"Обработка окон: обучение={train_window}, прогноз={predict_window} (wind={wind})")
        print("=" * 80)

        # Список папок с Excel‑файлами для этой пары окон
        folder_paths = [
            rf"Результаты_IMM-EKF_0.7_{wind}"
        ]

        # Перебираем папки
        for folder_path in folder_paths:
            results = []

            # перебор всех xlsx в текущей папке
            for file_path in glob(os.path.join(folder_path, "*.xlsx")):
                df = pd.read_excel(file_path)

                missing_cols = [col for col in col_names if col not in df.columns]
                if missing_cols:
                    print(f"В файле {file_path} нет столбцов: {missing_cols}")
                    continue

                means = df[col_names].mean(skipna=True)

                results.append({
                    "Имя файла": os.path.basename(file_path),
                    "средн % откл S": means["% откл S"],
                    "средн % откл I": means["% откл I"],
                    "средн % откл R": means["% откл R"],
                })

            if not results:
                print(f"В папке {folder_path} нет подходящих файлов или столбцов")
                continue

            result_df = pd.DataFrame(results)

            # Имя выходного файла по имени папки + окнам
            folder_name = os.path.basename(folder_path.rstrip(r"\/"))
            out_name = f"итог_средние_{folder_name}.xlsx"

            # Сохраняем в рабочей директории (где выполняется код)
            out_path = os.path.join(work_dir, out_name)
            result_df.to_excel(out_path, index=False)

            print(f"Готово для папки {folder_name}: {out_path}")
```
